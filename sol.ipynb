```python
import numpy as np
import pandas as pd
```

### DATA

```python
movies = pd.read_csv("datasets/tmdb_5000_movies.csv")
credits = pd.read_csv("datasets/tmdb_5000_credits.csv")
```

### PREPROCESSING

```python
movies.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.info()

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# ['Action' , Adventure , Fantasy , scifi]

#its your string of list we have to convert this into list 
# import ast (and write ast.litral_eval(obj) in place of only obj )
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# movies.head()
```

```python
def convert_top_5(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 5:  # Changed from 3 to 5
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

```

```python
movies['cast']=movies['cast'].apply(convert_top_5)
```

```python
# movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# movies.head()
```

```python
movies['overview'][0]
# its a string we wanna convert this in the list so that we can contatinate with other lists
# the function name is lamda
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
# UPDATE THIS CELL IN YOUR NOTEBOOK
new_df = movies[['movie_id', 'title', 'tags', 'genres', 'cast', 'crew']]

# Keep your existing code underneath it:
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
# Text to vector conversion 
# we will recommend those movies which has lower angle with movie is on going
# it called text vectorization
# type of text vectorization
# we gonna use bag of words.(other ar - tfidf , word2vec)
```

```python
# for words like love, loved, loving etc we will use stemming (PorterStemmer) to convert all these words into one word (love) so that we can get better results
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
feature_names = cv.get_feature_names_out()
print(f"Total features: {len(feature_names)}")
print('\n'.join(feature_names))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
